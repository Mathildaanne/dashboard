## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Mapping season ke nama musim
season_mapping = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
day_df["season"] = day_df["season"].map(season_mapping)

# Hitung rata-rata penyewaan per musim
seasonal_rentals = day_df.groupby("season")["cnt"].mean()

# Buat diagram batang yang lebih mudah dibaca
plt.bar(seasonal_rentals.index, seasonal_rentals.values, color=["green", "orange", "brown", "blue"])

# Tambahkan label
plt.xlabel("Musim")
plt.ylabel("Rata-rata Penyewaan")
plt.title("Penyewaan Sepeda Berdasarkan Musim")
plt.xticks(rotation=0)  # Pastikan label tidak miring
plt.grid(axis="y", linestyle="--", alpha=0.5)  # Tambahkan garis bantu

# Tampilkan grafik
plt.show()
```

### Pertanyaan 2:

```python
hourly_rentals = hour_df.groupby("hr")["cnt"].mean()

plt.figure(figsize=(10,5))
plt.plot(hourly_rentals.index, hourly_rentals.values, marker="o", linestyle="-", color="blue")

plt.xlabel("Jam dalam Sehari (0-23)")
plt.ylabel("Rata-rata Penyewaan")
plt.title("Tren Penyewaan Sepeda Berdasarkan Jam")
plt.xticks(range(0, 24))
plt.grid(True)

plt.show()
```

## Analisis Lanjutan (Opsional)

```python
rfm_df = day_df.groupby("dteday", as_index=False).agg({
    "dteday": "max",
    "cnt": "sum"
})

rfm_df.columns = ["date", "monetary"]

rfm_df["date"] = pd.to_datetime(rfm_df["date"])

recent_date = rfm_df["date"].max()

rfm_df["recency"] = rfm_df["date"].apply(lambda x: (recent_date - x).days)

rfm_df["frequency"] = 1
rfm_df.drop("date", axis=1, inplace=True)

print(rfm_df.head())
```

```python
rfm_sorted_frequency = rfm_df.sort_values(by="frequency", ascending=False).head(5)

rfm_sorted_monetary = rfm_df.sort_values(by="monetary", ascending=False).head(5)

fig, ax = plt.subplots(1, 2, figsize=(15, 6))
colors = sns.color_palette("viridis", 5)

sns.barplot(
    y="frequency",
    x=rfm_sorted_frequency.index,
    data=rfm_sorted_frequency,
    hue=rfm_sorted_frequency.index,
    palette=colors,
    legend=False,
    ax=ax[0]
)
ax[0].set_title("Top 5 Berdasarkan Frequency", loc="center", fontsize=18)
ax[0].set_xlabel("Date")
ax[0].set_ylabel("Frequency")
ax[0].tick_params(axis='x', labelsize=10)

sns.barplot(
    y="monetary",
    x=rfm_sorted_monetary.index,
    data=rfm_sorted_monetary,
    hue=rfm_sorted_monetary.index,
    palette=colors,
    legend=False,
    ax=ax[1]
)
ax[1].set_title("Top 5 Berdasarkan Monetary", loc="center", fontsize=18)
ax[1].set_xlabel("Date")
ax[1].set_ylabel("Monetary")
ax[1].tick_params(axis='x', labelsize=10)

plt.suptitle("Top 5 Hari Berdasarkan Parameter RFM", fontsize=20)

plt.tight_layout()
plt.show()
```
